### Ejercicio 2

Se quiere calcular la raíz de una función continua en un intervalo $[a, b]$ mediante el método
de bisección, de forma tal que el error cometido no exceda una tolerancia prefijada denotada por $\varepsilon$. Es decir, si $x_n$ es el último valor calculado se debe satisfacer $|x_n − x_∗| ≤ \varepsilon$, donde
$x_∗$ es la raíz en cuestión.

* a) Estimar el número de iteraciones $k$ que son suficientes para satisfacer este requerimiento en términos de la tolerancia $\varepsilon$ y la longitud del intervalo $[a, b]$.

* b) ¿Cuántas iteraciones resultan si $[a, b] = [0, 1]$ y $ε = 10^{−5}$?.

a) Por un teorema, sabemos que el método de bisección genera una sucesión $\{x_n\}_{n=1}^∞$ que converge a la raíz de $x_*$ de $f$ tal que:

$$
|x_*-x_n| \leq \frac{1}{2^{n+1}}(b-a)
$$

Nos interesa para este ejercicio, enocntrar un $k$ tal que dada la tolerancia $\varepsilon$, se tenga que:

$$
|x_*-x_k| \leq \frac{1}{2^{k+1}}(b-a)\leq\varepsilon
$$

Para ello despejamos:

\begin{align*}
\frac{1}{2^{n+1}}(b-a)\leq\varepsilon &⇔ 2^{n+1}\geq\frac{b-a}{\varepsilon}\\
&⇔n+1\geq \log_2\left(\frac{b-a}{\varepsilon}\right)
\end{align*}

elegimos $k$ como el primer natural tal que $\log_2\left(\frac{b-a}{\varepsilon}\right)\leq k+1$, es decir, elegimos a $k$ como la parte entera de $\log_2\left(\frac{b-a}{\varepsilon}\right)$

$$
k=\left[\log_2\left(\frac{b-a}{\varepsilon}\right)\right]
$$



b) Para $\varepsilon = 10^{-5}$ y $[a,b]=[0,1]$ tenemos

$$
k=\left[\log_2\left(\frac{1}{10^{-5}}\right)\right]= \left[\log_2\left(10^5\right)\right]=[5\log_2(10)]=16
$$

Por lo que se necesita  al menos $16$ iteracciones.

## Método de Newton

### Ejercicio 4

Dado $a>0$, para calcular $\sqrt{a}$ se debe hallar la raíz de $f(x)=x^2-a$.

* a) Mostrar que la iteración de Newton genera la siguiente sucesión: $x_{n+1}=\frac{1}{2}\left(x_n+\frac{a}{x_n}\right)$.

* b) Probar que para cualquier $x_0\in(0,∞)$, las aproximaciones generadas por el método de Newton satisfacen $x_n\geq \sqrt{a}$ para $n\geq 1$.

* c) Probar que la sucesión es decreciente $(x_n\geq x_{n+1}$ para todo $n\geq 1$).

* d) Finalmente concluir que la sucesión generada por el algoritmo converge a $\sqrt{a}$.

a) Tenemos:

\begin{align*}
x_{n+1}&=x_n-\frac{f(x_n)}{f'(x_n)}=x_n-\frac{x_n^2-a}{2x_n}\\
&=\frac{2x_n^2}{2x_n}-\frac{x_n^2-a}{2x_n}\\ &= \frac{1}{2}\frac{2x_n^2-x_n^2-a}{x_n} \\ &=\frac{1}{2}\frac{x_n-a}{x_n}\\ &=\frac{1}{2}\left(1-\frac{a}{x_n}\right)
\end{align*}

como queríamos probar.

b) La interpretación gráfica de lo que queremos demostrar es la siguiente

image.png

En efecto:

\begin{align*}
x_n\geq\sqrt{a} &⇔ \frac{1}{2}\left(x_{n-1}+\frac{a}{x_{n-1}}\right)\geq\sqrt{a}\\
&⇔x_{n-1}+\frac{a}{x_{n-1}}\geq 2\sqrt{a} \\
&⇔x_{n-1}+\frac{a}{x_{n-1}}-2\sqrt{a}\geq 0 \\
&⇔\frac{x_{n-1}^2+a-2\sqrt{a}x_{n-1}}{x_{n-1}}\geq 0\\
&⇔x_{n-1}^2+a-2\sqrt{a}x_{n-1}\geq0\\
&⇔\left(x_{n-1}-\sqrt{a}\right)^2\geq0
\end{align*}

Como la última expresión es cierta para todo $n$, entonces $x_n\geq\sqrt{a}$ es cierto para todo $n$, bajo las hipótesis que plantea el inciso.

c) Queremos probar que para todo $x_n\geq x_{n+1}$ para todo $n\geq1$. En efecto

\begin{align*}
x_n\geq x_{n+1} &⇔ x_n \geq \frac{1}{2}\left(x_n+\frac{a}{x_n}\right)\\
&⇔2x_n\geq x_n + \frac{a}{x_n} \\
&⇔x_n\geq \frac{a}{x_n} \\
&⇔x_n^2\geq a
\end{align*}

En la última expresión podemos tomar raíz cudrada a ambos lados de la desigualdad ya que por hipótesis $x_n,a>0$. Con lo cual la desigualdad es cierta si y sólo si $x_n\geq\sqrt{a}$, algo que probamos en b).

d) Por el ítem b), la sucesión es acotada inferiormente por $\sqrt{a}$ y por el ítem e) la sucesión es decreciente, de ahí se deduce que la sucesión es convergente y converge a $\sqrt{a}$ (teorema de Análisis 1).

### Ejercicio 5

Proponer una fórmula de iteración para aproximar el recíproco de la raiz cuadrada de un número positivo $r$ utilizando el Método de Newton. Decidir cuántas iteraciones son necesarias para que el error relativo de dicha aproximación sea menor que $10^{−4}$ partiendo de $x_0 = 1$ y considerando $r = 5$.

Queremos hallar una fórmula de iteración para, dado un $r$ positivo, calcular la raíz cuadrada de su recíproco:

$$
\frac{1}{\sqrt{r}}
$$

Para ello, podemos buscar un $x_*$ que sea raíz de la función:

$$
f(x)=x^2-\frac{1}{r}
$$

Pues si eso pasa entonces:

\begin{align*}
x_*^2-\frac{1}{r}=0 &⇔x_*^2=\frac{1}{r} \\
&⇔x_*=\frac{1}{\sqrt{r}}
\end{align*}

Utilizamos ahora el método de Newton para hallar la fórmula iterativa:

\begin{align*}
x_{n+1}=x_n-\frac{f(x)}{f'(x)} &= x_n -\frac{x_n^2+\frac{1}{r}}{2x_n}\\
&=\frac{2x^2_n-x^2_n+\frac{1}{r}}{2x_n}\\
&=\frac{x_n^2+\frac{1}{r}}{2x_n} \\
&= \frac{1}{2}\left(x_n+\frac{1}{rx_n}\right)
\end{align*}

Vamos ahora a iterar hasta que el error relativo $\delta_{\text{rel}}$ sea menor a $10^{-4}$. Usamos $x_0=1$ y $r=5$ como la aproximación inicial, luego para la primer iteración tenemos:

\begin{align*}
\delta_{\text{rel}_1}=\frac{\left|\frac{1}{\sqrt{5}}-x_1\right|}{\left|\frac{1}{\sqrt{5}}\right|} &= \frac{\left|\frac{1}{\sqrt{5}}-\frac{1}{2}\left(x_0+\frac{1}{rx_0}\right)\right|}{\left|\frac{1}{\sqrt{5}}\right|}\\
&=\frac{\left|\frac{1}{\sqrt{5}}-\frac{1}{2}\left(1+\frac{1}{5}\right)\right|}{\left|\frac{1}{\sqrt{5}}\right|}\\
&=\frac{\left|\frac{1}{\sqrt{5}}-\frac{3}{5}\right|}{\left|\frac{1}{\sqrt{5}}\right|}≈0.34
\end{align*}

Que no es del orden de $10^{-4}$. Para la segunda iteración, teniendo en cuenta que $x_1=\frac{3}{5}$.

\begin{align*}
\delta_{\text{rel}_2}=\frac{\left|\frac{1}{\sqrt{5}}-x_2\right|}{\left|\frac{1}{\sqrt{5}}\right|} &=\frac{\left|\frac{1}{\sqrt{5}}-\frac{1}{2}\left(x_1+\frac{1}{rx_1}\right)\right|}{\left|\frac{1}{\sqrt{5}}\right|}\\
&= \frac{\left|\frac{1}{\sqrt{5}}-\frac{1}{2}\left(\frac{3}{5}+\frac{5}{15}\right)\right|}{\left|\frac{1}{\sqrt{5}}\right|}\\
&=\frac{\left|\frac{1}{\sqrt{5}}-\frac{7}{15}\right|}{\left|\frac{1}{\sqrt{5}}\right|} ≈ 0.04
\end{align*}

Todavía no cumple con el error buscado. Hagamos una iteración más:

\begin{align*}
\delta_{\text{rel}_3}=\frac{\left|\frac{1}{\sqrt{5}}-x_3\right|}{\left|\frac{1}{\sqrt{5}}\right|} &=\frac{\left|\frac{1}{\sqrt{5}}-\frac{1}{2}\left(x_2+\frac{1}{rx_2}\right)\right|}{\left|\frac{1}{\sqrt{5}}\right|}\\
&= \frac{\left|\frac{1}{\sqrt{5}}-\frac{1}{2}\left(\frac{7}{15}+\frac{15}{7\cdot 5}\right)\right|}{\left|\frac{1}{\sqrt{5}}\right|}\\
&=\frac{\left|\frac{1}{\sqrt{5}}-\frac{94}{210}\right|}{\left|\frac{1}{\sqrt{5}}\right|} ≈ 0.0009 < 10^{-4}
\end{align*}

Con lo cual son necesarias tres iteraciones.

b) Mostrar que si $\{x_n\}$ es la sucesión generada por el método de Newton para la función $f(x)=\text{arctan}(x)$ con $x_0=α$, se cumple que $x_n=(-1)^nα$.

a) Veamos que se cumplan las hipótesis del teorema del valor medio. Por un lado $g$ es derivable para todo $x$:

$$
g'(x)=\frac{1}{1+x^2}-\frac{2(1-x^2)}{(1+x^2)^2}
$$

Con lo cual es continua para todo $x$. Por otro lado:

* $g(1)=\frac{π}{4}-1<0$
* $g(\sqrt{3})=\frac{π}{3}-\frac{\sqrt{3}}{2}>0$
* ⇒ $g(1)<g(\sqrt{3})$ y $0\in(g(1),g(\sqrt{3}))$

El teorema del valor medio nos dice que para todo $u\in(g(1),g(\sqrt{3}))$ existe $r$ tal que $g(r)=u$. En particular, para $0\in(g(1),g(\sqrt{3}))$ existe $α$ tal que $g(α)=0$.



b) Por un lado:

\begin{align*}
g(α)=0&⇒\text{arctan}(α)-\frac{2α}{1+α^2}=0\\
&⇒\text{arctan}(α)=\frac{2α}{1+α^2}
\end{align*}

Por otro lado:

\begin{align*}
x_{n+1}&=x_n-\frac{g(x_n)}{g'(x_n)} \\
&=x_n-\frac{\text{arctan}(x_n)}{\frac{1}{1+x_n^2}}\\
&=x_n -(1+x_n^2)\text{arctan}(x_n)
\end{align*}

Si iteramos, con aproximación inicial $x_0=α$, teniendo en cuenta que al ser $\text{arctan}(x)$ impar entonces $\text{arctan}(-α)=-\text{arctan}(α)$ tenemos:

\begin{align*}
x_0&=α\\
x_1&=x_0 -(1+x_0^2)\text{arctan}(x_0) \\&= α -(1+α^2)\text{arctan}(α)\\&=α -(1+α^2)\frac{2α}{1+α^2} \\&=α-2α\\&=-α \\
x_2&=x_1 -(1+x_1^2)\text{arctan}(x_1)\\ &= -α -(1+(-α)^2)\text{arctan}(-α)\\&= -α +(1+(-α)^2)\frac{2α}{1+α^2} \\ &= -α+2α\\&= α\\
&\hspace{0.3cm}\vdots\\
x_n&=(-1)^nα
\end{align*}

Se desea hallar la mínima raíz positiva de las siguientes funciones mediante el método de
iteración de punto fijo. Para cada una de ellas determinar una función de iteración y un

a) $f(x)=x^3-x-1$

Buscamos $x_*\in(0,∞)$ tal que $f(x_*)=0$, o sea:

\begin{align*}
x_*^3-x_*-1=0
\end{align*}



Si despejamos $x_*$, obtenemos dos funciones:

$$
g_1(x)=x^3-1 \\
g_2(x)=\sqrt[3]{x+1}
$$

Lo que tenemos ahora es el problema de punto fijo de encontrar $p$ tal que $g_1(p)=p$ o $g_2(p)=p$. Vamos a usar el teorema del punto fijo para encontrar dichos $p$.




Para $g_1$ tenemos:

$$
g_1(x)=x^3-1⇒ g_1'(x)=3x^2
$$

Buscamos $k$ tal que $|g_1'|\leq k < 1$. Tenemos:

\begin{align*}
|3x^2| \leq k ⇔ |x|\leq \sqrt{\frac{k}{3}}<\sqrt{\frac{1}{3}}
\end{align*}

Es decir, $|g_1'(x)|< 1$ si $x\in(-\sqrt{1/3}, \sqrt{1/3})$.

La raíz que buscamos es positiva, con lo cual el intervalo donde buscamos un punto fijo para $g_1$ es subconjunto cerrado de $\left(0,\frac{1}{\sqrt{3}}\right)$. Sin embargo, $g_1$ es creciente debido a que $g'_1\geq0$, $g_1(0)=-1$ y $g\left(\frac{1}{\sqrt{3}}\right)<0$. Es decir $g_1<0$ en cualquier subintervalo cerrado de  $\left(0,\frac{1}{\sqrt{3}}\right)$, y dicho intervalo es estrictamente positivo.

Esto nos dice que no se satisface la hipótesis para la existencia de un punto fijo: no existe $[a,b]⊂(0,1/\sqrt{3})$ tal que $g_1(x)\in[a,b]$.


Para $g_2$, que es creciente, tenemos:

$$
g_2'(x)=\frac{(x+1)^{-\frac{2}{3}}}{3}
$$

que es continua, y:

\begin{align*}\tag{$x>0$}
\left|\frac{(x+1)^{-\frac{2}{3}}}{3}\right| \leq k &⇔\frac{(x+1)^{-\frac{2}{3}}}{3} \leq k\\
&⇔(x+1)^{-2/3}\leq 3k\\
&⇔x+1\geq (3k)^{-3/2}\\
&⇔x\geq (3k)^{-3/2}-1\\ \tag{$k<1$}
&⇔x> (3)^{-3/2}-1\\
&⇔x>\frac{1}{\sqrt{27}}-1
\end{align*}

Con lo cual, el intervalo donde debemos buscar $p$ es

$$
I=[a,b]\subset\left(\frac{1}{\sqrt{27}}-1,\infty\right)
$$

Como buscamos la raíz positiva, entonces queremos $x>0$ y $g_2(x)>0$. Observemos que $g_2(0)=1$ con lo cual podemos tomar $I=[1,b]$.

Para determinar $b$, dado que $g_2(x)$ es creciente, planteamos:

$$
g(b)\leq b⇔\sqrt[3]{b+1}\leq b
$$

Para $b=2$ tenemos:

$$
g(2)=\sqrt[3]{2+1}=\sqrt[3]{2}\leq 2
$$

Con lo cual podemos tomar $b=2$, y tenemos $1\leq g(1)=\sqrt[3]{2}<g(2)\leq 2$. En resumen, hemos obtenido:

* Una función $g=g_2$ de iteración.

* Un intervalo $I=[1,2]$ donde la función es continua y $ x\in(1,2)⇒ g\in(1,2)$

* $g'(x)$ existe y se cumple $|g'(x)|<1$ para $x\in I$

Con lo cual el teorema del punto fijo nos dice que la función de iteración $g$ iniciada en un punto $p_0\in I$ convergerá a un único punto fijo $p$ en $I$, que será una raíz de $f$.

Además como construimos $I$ para que este contenga una raíz positiva, entonces $p$, al ser única, es la primer raíz positiva de $f$ (en el intervalo $[0,1)$, $g(x)<0$, no hay raíz en dicho intervalo)
